# 03. 도형 그리기

## 빈 스케치북 만들기

```python
import cv2
import numpy as np
```

```python
# 세로 480 X 가로 640, 3 Channel (RGB)에 해당하는 스케치북 만들기
# OpenCV에서는 RGB가 아닌 BGR로 사용된다.

img = np.zeros((480, 640, 3), dtype=np.uint8)  # 모두 0으로 채움
# img[:] = (255, 255, 255)  # (B,G,R) 전체 공간을 흰 색으로 채우기
# print(img)

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 일부 영역 색칠

```python
img = np.zeros((480, 640, 3), dtype=np.uint8)

# img[세로 시작 : 세로 끝, 가로 시작: 가로 끝]
img[100:200, 200:300] = (255, 255, 255)

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 직선
직선의 종류(line type)

1. cv2.LINE_4 : 상하 좌우 4방향으로 연결된 선
2. cv2.LINE_8 : 대각선을 포함함 8방향으로 연결된 선 (기본값)
3. cv2.LINE_AA : 부드러운 선 (anti-aliasing)

```python
img = np.zeros((480, 640, 3), dtype=np.uint8) 

COLOR = (0, 255, 255) # BGR : YELLOW, 색깔
THICKNESS = 3 # 두께

#cv2.line(그릴위치(= 도화지), 시작점(x, y), 끝점(x, y), 색, 두께, 선 타입)
cv2.line(img, (50, 100), (400, 50), COLOR, THICKNESS, cv2.LINE_8)
cv2.line(img, (50, 200), (400, 150), COLOR, THICKNESS, cv2.LINE_4)
cv2.line(img, (50, 300), (400, 250), COLOR, THICKNESS, cv2.LINE_AA)

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 원

```python
img = np.zeros((480, 640, 3), dtype=np.uint8) 

COLOR = (255, 255, 0) # BGR : 옥색
RADIUS = 50 # 반지름
THICKNESS = 10 # 두께

# cv2.circle(그릴위치(= 도화지), 원의중심점(x, y), 반지름, 색, 두께, 선 타입)
cv2.circle(img, (200, 100), RADIUS, COLOR, THICKNESS, cv2.LINE_AA) # 속이 빈 원
cv2.circle(img, (400, 100), RADIUS, COLOR, cv2.FILLED, cv2.LINE_AA) # 속이 꽉 찬 원 | cv2.FILLED : -1로 작성해도 됨

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 사각형

```python
img = np.zeros((480, 640, 3), dtype=np.uint8) 

COLOR = (0, 255, 0) # BGR : 초록색
THICKNESS = 3 # 두께

# cv2.rectangle(그릴위치(= 도화지), 왼쪽 위 좌표(x, y), 오른쪽 아래 좌표(x, y), 색, 두께)
cv2.rectangle(img, (100, 100), (200, 200), COLOR, THICKNESS) # 속이 빈 사각형
cv2.rectangle(img, (300, 100), (400, 300), COLOR, cv2.FILLED) # 속이 꽉 찬 사각형

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 다각형

```python
img = np.zeros((480, 640, 3), dtype=np.uint8) 

COLOR = (0, 0, 255) # BGR : 빨간색
THICKNESS = 3 # 두께

pts1= np.array([[100, 100],[200, 100],[100, 200]])
pts2= np.array([[200, 100],[300, 100],[300, 200]])

# cv2.polylines(img, 그릴 좌표들, 닫힘 여부, 색, 두께, 선 타입)

# [속이 빈 다각형]
# cv2.polylines(img, [pts1], True, COLOR, THICKNESS, cv2.LINE_AA)
# cv2.polylines(img, [pts2], True, COLOR, THICKNESS, cv2.LINE_AA)
cv2.polylines(img, [pts1, pts2], True, COLOR, THICKNESS, cv2.LINE_AA) 

# [꽉 찬 다각형]
pts3 = np.array([[[100, 300],[200, 300],[100, 400]], [[200, 300],[300, 300],[300, 400]]])
cv2.fillPoly(img, pts3, COLOR, cv2.LINE_AA) # 속을 채우기 때문에 닫힘여부, 선 두께는 작성할 필요가 없음
# pts3은 리스트로 감싸져있기에 좌표 작성시 리스트로 감싸지 않음. => pts3(o), [pts3](x)

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
